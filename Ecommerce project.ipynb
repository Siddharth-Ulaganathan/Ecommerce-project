6. Highest and Lowest Purchase Prices.

```python
print("Highest price:\n",df['Purchase Price'].max(),"\nLowest price:\n",df['Purchase Price'].min())
```

7. Average Purchase Price

```python
print("Average price:\n", df['Purchase Price'].mean())
```

8. How many people have French 'fr' as their Language?

```python
print(df['Language'].str.contains('fr').sum())
```

9. Job Title Contains Engineer

```python
print(df['Job'].str.contains('engineer|Engineer').sum())
```

10. Find The Email of the person with the following IP Address: 132.207.160.22

```python
print(df.loc[df['IP Address']=='132.207.160.22','Email'])
```

11. How many People have Mastercard as their Credit Card Provider and made a purchase above 50?

```python
print((df['CC Provider'].str.contains('Mastercard')&(df['Purchase Price']>50)).sum())
```

12. Find the email of the person with the following Credit Card Number: 4664825258997302

```python
print(df.loc[df['Credit Card']==4664825258997302,'Email'])
```

13. How many people purchase during the AM and how many people purchase during PM?

```python
print(f"People purchases during AM:\n{(df['AM or PM']=='AM').sum()}\n",f"People purchases during PM:\n{(df['AM or PM']=='PM').sum()}")
```

14. How many people have a credit card that expires in 2020?

```python
print(df['CC Exp Date'].str.endswith('20').sum())
```

15. What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)

```python
print(df["Email"].str.split('@').str[1].value_counts().head())
```

```python
print(df.groupby(df['Email'].str.split('@').str[1]).size().sort_values(ascending=False).head())
```
